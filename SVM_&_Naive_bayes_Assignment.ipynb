**1. What is a Support Vector Machine (SVM)?**

SVM is a supervised machine learning algorithm used for classification and regression tasks. It finds the optimal hyperplane that best separates data points of different classes in a high-dimensional space.

**2.  What is the difference between Hard Margin and Soft Margin SVM?**

Hard Margin SVM assumes data is perfectly linearly separable and does not allow misclassification.

Soft Margin SVM introduces a penalty for misclassified points, allowing better generalization when data is noisy or overlapping.


**3. What is the mathematical intuition behind SVM?**

SVM aims to maximize the margin between the closest data points (support vectors) and the decision boundary. This leads to better generalization by minimizing structural risk.

**4. What is the role of Lagrange Multipliers in SVM?**

Lagrange Multipliers are used to solve the constrained optimization problem in SVM. They help convert the primal problem into a dual form, making it easier to handle high-dimensional data and apply the kernel trick.


**5.What are Support Vectors in SVM?**

Support Vectors are the data points closest to the decision boundary. They are critical in defining the position and orientation of the hyperplane.

**6. What is a Support Vector Classifier (SVC)?**

SVC is the classification implementation of SVM. It uses support vectors to classify data into categories by finding the optimal separating hyperplane.

**7 What is a Support Vector Regressor (SVR)?**

SVR is the regression counterpart of SVM. It tries to fit the data within a margin of tolerance and penalizes deviations outside this margin.

**8. What is the Kernel Trick in SVM?**

The Kernel Trick allows SVM to operate in a high-dimensional space without explicitly computing the coordinates. It transforms non-linearly separable data into a space where a linear separator is possible.

**9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?**
- Linear Kernel: Best for linearly separable data; fast and simple.
- Polynomial Kernel: Captures interactions between features; good for medium complexity.
- RBF Kernel: Handles complex, non-linear relationships; most commonly used.

**10. What is the effect of the C parameter in SVM?**

The C parameter controls the trade-off between maximizing the margin and minimizing classification error. A small C allows a wider margin with more misclassifications; a large C prioritizes correct classification over margin width.

**11. What is the role of the Gamma parameter in RBF Kernel SVM?**

Gamma defines how far the influence of a single training example reaches. Low gamma means far reach (smoother decision boundary), high gamma means close reach (more complex boundary).

**12. What is the Naïve Bayes classifier, and why is it called "Naïve"?**
Naïve Bayes is a probabilistic classifier based on Bayes’ Theorem. It’s called "naïve" because it assumes feature independence, which is rarely true in real-world data.

**13. What is Bayes’ Theorem?**

Bayes’ Theorem describes the probability of an event based on prior knowledge of conditions related to the event:
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}


**14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?**

- Gaussian NB: Assumes features follow a normal distribution; used for continuous data.
- Multinomial NB: Assumes features are counts; ideal for text classification.
- Bernoulli NB: Assumes binary features; suitable for binary/boolean data.

**15. When should you use Gaussian Naïve Bayes over other variants?**

Use Gaussian NB when your features are continuous and approximately normally distributed, such as in sensor data or medical measurements.


**16. What are the key assumptions made by Naïve Bayes?**

- Features are conditionally independent given the class.
- Each feature contributes equally and independently to the outcome.

**17. What are the advantages and disadvantages of Naïve Bayes?**

Advantages: Fast, simple, works well with high-dimensional data, especially text.
Disadvantages: Assumes independence, which may not hold; can perform poorly with correlated features.

**18. Why is Naïve Bayes a good choice for text classification?**

Naïve Bayes handles high-dimensional sparse data efficiently, assumes word independence (reasonable in many text tasks), and performs well with small training sets.

**19. Compare SVM and Naïve Bayes for classification tasks?**

Support Vector Machine (SVM) is a discriminative algorithm that focuses on finding the optimal boundary (hyperplane) between classes. It works well with complex and high-dimensional data, especially when the decision boundary is not obvious. SVM is powerful but computationally intensive and sensitive to parameter tuning.


Naïve Bayes, on the other hand, is a generative model that calculates the probability of each class using Bayes’ Theorem and assumes that features are conditionally independent. It’s extremely fast, easy to implement, and performs surprisingly well on text classification tasks, even when the independence assumption is violated.
In short, SVM is ideal for precision and complex boundaries, while Naïve Bayes is preferred for speed and simplicity, especially in text-heavy domains.


**20. How does Laplace Smoothing help in Naïve Bayes?**

Laplace Smoothing helps Naïve Bayes avoid assigning zero probability to features that were not seen in the training data. Without smoothing, if a word or feature never appears in a class during training, its probability becomes zero, which can cause the entire class prediction to fail.

To fix this, Laplace Smoothing adds a small constant (usually 1) to every feature count. This ensures that every possible feature has at least a minimal probability, making the model more robust and better at handling unseen data during prediction. It's especially useful in text classification where rare words are common

```python
#24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
#boundary:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC

X, y = make_moons(n_samples=200, noise=0.2, random_state=42)
model = SVC(kernel='poly', degree=3)
model.fit(X, y)

# Plot decision boundary
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title("SVM with Polynomial Kernel")
    plt.show()

plot_decision_boundary(model, X, y)
```

```python
#27. Write a Python program to train an SVM Classifier with different C values and compare the decision
#boundaries visually?
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC

X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=42)

def plot_svm(C_value):
    model = SVC(kernel='linear', C=C_value)
    model.fit(X, y)

    plt.figure()
    plt.title(f"SVM Decision Boundary (C={C_value})")
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.show()

for C in [0.01, 1, 100]:
    plot_svm(C)
```

```python
#38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy?
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np

X, y = load_iris(return_X_y=True)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracies = []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = SVC()
    model.fit(X_train, y_train)
    acc = accuracy_score(y_test, model.predict(X_test))
    accuracies.append(acc)

print("Average Accuracy:", np.mean(accuracies))
```

```python
#43. Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn?

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix for SVM Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score?

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = GaussianNB()
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]

print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
#46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve?
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC(probability=True)
model.fit(X_train, y_train)
y_scores = model.predict_proba(X_test)[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_scores)
disp = PrecisionRecallDisplay(precision=precision, recall=recall)
disp.plot()
plt.title("Precision-Recall Curve for SVM Classifier")
plt.show()
```
